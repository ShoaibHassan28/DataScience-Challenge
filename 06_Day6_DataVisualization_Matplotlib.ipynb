## ðŸ“Œ What is Data Visualization
Data Visualization is the graphical representation of data using plots and charts to identify patterns, trends, and insights.

```python
import matplotlib.pyplot as plt
import numpy as np
```

# ðŸ“Œ Line Plot

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# ðŸ“Œ Bar Plot 

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 9]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Plot Example")
plt.show()
```

# ðŸ“Œ Scatter Plot

```python
x = np.array([5, 7, 8, 7, 2, 17, 2, 9])
y = np.array([99, 86, 87, 88, 100, 86, 103, 87])

plt.scatter(x, y, color='red')
plt.title("Scatter Plot Example")
plt.show()
```

# ðŸŽ¯ Mini Task
## Create:

A line plot of squares of numbers 1 to 10

A bar plot of 4 products with sales values




```python

numbers = np.arange(1, 11)
squares = numbers ** 2

plt.plot(numbers, squares, marker='o')
plt.title("Line Plot: Squares of Numbers 1 to 10")
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid(True)
plt.show()

products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 120, 300]

plt.bar(products, sales, color='orange')
plt.title("Bar Plot: Sales of 4 Products")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

```
